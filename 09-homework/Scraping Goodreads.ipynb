````python
#find the chunk
books = allbooks.find_all("tr", itemscope_="")
books
````

````python
# not the cleaned up version
rows = []
for book in books:
    listing = {}
    rank = book.find("td", class_="number").text.strip()
    title = book.find(class_="bookTitle").text.strip()
    author = book.find("a",class_="authorName").text.strip()
    
    # i am sure there is a more efficient way
    scores = book.find_all("span", class_="smallText")
    for score in scores:
        stars = score.find("a")
    
    vote = book.find("img", class_="loading").previous    
    rating = book.find("span", class_="minirating").text    
    print(rank,"|",title,"|",author,"|",stars.text,"|",vote,"|",rating)
    listing = {
        'rank':rank,
        'title':title,
        'author':author,
        'score':stars.text,
        'votes':vote,
        'rating':rating
    }
    print(listing)
    rows.append(listing)
rows
````

````python
df=pd.DataFrame(rows)
df.head()
````

````python
#don't use header!
df.to_csv("goodreads.csv", index=False)
````
